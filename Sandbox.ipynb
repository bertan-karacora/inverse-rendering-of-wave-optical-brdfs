 # WaveOpticsBRDF sandbox

## Setup

### Libraries

```python
%%bash
apt install libeigen3-dev
apt install openexr
```

### Download code

```python
!git clone "https://github.com/pixelrein-bertan/Inverse-Rendering-Inverse-Rendering-for-Capturing-Specular-Microgeometry-from-Wave-Optical-BRDFs.git"
%cd WaveOpticsBrdf
```

### Compile

```python
!make -f makefile.linux
```

### Download heightfields

```python
%%bash
mkdir heightfields
wget -O heightfields/isotropic.exr https://www.dropbox.com/s/siepjp35pfw218i/isotropic.exr?dl=1
wget -O heightfields/scratched.exr https://www.dropbox.com/s/siepjp35pfw218i/scratched.exr?dl=1
wget -O heightfields/brushed_8K.exr https://www.dropbox.com/s/siepjp35pfw218i/brushed_8K.exr?dl=1
```

## Run

```python
%cd WaveOpticsBrdf
!./genBrdf -m Wave -d OHS -i heightfields/isotropic.exr -o isotropicBrdf.exr -x 512 -y 512 -p 10.0 -w 1.0 -s -1.0 -t -1.0 -r 128
!./genBrdf -m Wave -d OHS -i heightfields/scratched.exr -o scratchedBrdf.exr -x 512 -y 512 -p 10.0 -w 1.0 -s -1.0 -t -1.0 -r 128
!./genBrdf -m Wave -d OHS -i heightfields/brushed_geom_brdf.exr -o brushed_geom_brdf.exr -x 512 -y 512 -p 10.0 -w 1.0 -s -1.0 -t -1.0 -r 128
```
